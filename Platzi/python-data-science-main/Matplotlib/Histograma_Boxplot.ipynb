# Gr치ficos de Histograma y Boxplot para distribuciones

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(170, 10, 200)
print(data)
```

```python
plt.hist(data, color='salmon', bins=20, edgecolor='black', alpha=0.6)
plt.title('Distribuci칩n de alturas')
plt.xlabel('Altura (cm)')
plt.ylabel('Densidad')
plt.show()
```

```python
np.random.seed(0)
ages = [np.random.normal(30, 5, 100),
        np.random.normal(40, 5, 100),
        np.random.normal(35, 5, 100)]
print(ages)
```

```python
plt.boxplot(ages, patch_artist=True, notch=True, vert=True,
            labels=['Grupo 1','Grupo 2','Grupo 3'])

plt.title('Distribuci칩n de edades por grupo')
plt.xlabel('Grupo')
plt.ylabel('Edad (a침os)')
plt.show()
```
